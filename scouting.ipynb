# BMW Project Scouting

```python
from pandas import read_csv
from datetime import datetime as dt


def split_last_column(column_str):
    parts = column_str.split(";")
    ns = int(parts[0].split("=")[1])
    s = parts[1].split("s=")[1].replace('"', "").split(".")
    return (ns, s)
```

```python
data = read_csv("data/data_01.csv", parse_dates=[0, 1, 7])
```

## More Scouting on last column

```python
column = data["data_key_val_string"]
column_new = column.map(split_last_column)
```
