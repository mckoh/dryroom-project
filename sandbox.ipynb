```python
from pandas import read_csv, concat

data_a = read_csv("data/standalone/ST070_AZ203_BT36_Temperature.csv", parse_dates=[0])
data_a.index = data_a.trigger_ts_utc
data_a.drop("trigger_ts_utc", axis=1, inplace=True)
data_a["value_ST070_AZ203_BT36_Temperature"] = data_a.measured_value

data_b = read_csv("data/standalone/ST070_AZ203_BT37_Temperature.csv", parse_dates=[0])
data_b.index = data_b.trigger_ts_utc
data_b.drop("trigger_ts_utc", axis=1, inplace=True)
data_b["value_ST070_AZ203_BT37_Temperature"] = data_b.measured_value

data = concat(
    [
        data_a.value_ST070_AZ203_BT36_Temperature,
        data_b.value_ST070_AZ203_BT37_Temperature
    ], axis=1
)
```

```python
data.plot()
```
